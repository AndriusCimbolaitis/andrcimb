# 3. KIEK NUO RYKLIŲ ATAKŲ NUKENTĖJO MOTERŲ IR KIEK VYRŲ

```python
duomenys['Lytis'].unique()
```

```python
duomenys['countas'] = 1
```

```python
moterys = duomenys.loc[duomenys['Lytis'] == 'F'].count()['countas']
vyrai = duomenys.loc[(duomenys['Lytis'] == 'M') | (duomenys['Lytis'] == 'M ')].count()['countas']

print(moterys + vyrai)
print(moterys,vyrai)
```

```python
plt.pie([moterys, vyrai], labels=['Moterys', 'Vyrai'], autopct='%.2f %%')

plt.title('Ryklių atakos 1845-2017')
plt.show()
```

# 4. AR TIKRAI DAUGIAUSIAI RYKLIŲ UŽPUOLIMŲ PRIE AUSTRALIJOS KRANTŲ?

```python
duomenys['Salis'].unique()
```

```python
top_salys = duomenys.groupby('Salis').count()['Case Number'].sort_values(ascending=False).head(5)
top_salys
```

```python
top_salys.plot(kind='barh')


plt.title('Daugiausiai ryklių atakų šių šalių pakrantėse')

plt.xlabel('Atakų skaičius')
plt.ylabel('Šalys')

plt.show()
```
